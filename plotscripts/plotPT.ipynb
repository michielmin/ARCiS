```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

```python
# define the directory where the run is stored
dir='/Users/michielm/HAT-P-12b/NIRSPEC_HST/final_TEATRO/'
```

```python
# If there was a retrieval run postprocessed with pew=.true. you can use the PT_limits file
with open(dir+'/PT_limits') as file:
    data1 = [[float(digit) for digit in line.split()] for line in file]

nsamples=len(data1)

mx=[]
mym3=[]
mym2=[]
mym1=[]
myp1=[]
myp2=[]
myp3=[]
my=[]
for i in range(0,nsamples):
    mx.append(data1[i][0])
    my.append(data1[i][4])
    mym3.append(data1[i][1])
    myp3.append(data1[i][7])
    mym1.append(data1[i][3])
    myp1.append(data1[i][5])
    mym2.append(data1[i][2])
    myp2.append(data1[i][6])
```

```python
# Plot the posterior distribution of the PT profile
fig, (ax) = plt.subplots(1, 1, sharex=True)

ax.set_yscale("log")
ax.set_xlim(200,1200)
ax.set_ylim(1e2,1e-10)

ax.fill_betweenx(mx, mym3, myp3, color=(0,0,1),zorder=120,alpha=0.33)
ax.fill_betweenx(mx, mym2, myp2, color=(0,0,1),zorder=130,alpha=0.33)
ax.fill_betweenx(mx, mym1, myp1, color=(0,0,1),zorder=140,alpha=0.33)
ax.plot(my, mx, color=(0,0,0.5),zorder=150)

ax.set_ylabel(r"$P [\rm{bar}]$")
ax.set_xlabel(r"$T [\rm{K}]$")
```

```python
# To plot a simple PT profile from one forward model
with open(dir+'/mixingratios.dat') as file:
    data1 = [[digit for digit in line.split()] for line in file]

nsamples=len(data1)

mx=[]
my=[]
for i in range(1,nsamples):
    mx.append(float(data1[i][1]))
    my.append(float(data1[i][0]))
```

```python
# Plot a simple line of the PT profile
fig, (ax) = plt.subplots(1, 1, sharex=True)

ax.set_yscale("log")
ax.set_xlim(200,1200)
ax.set_ylim(1e2,1e-10)

ax.plot(my, mx, color=(0,0,0.5),zorder=150)

ax.set_ylabel(r"$P [\rm{bar}]$")
ax.set_xlabel(r"$T [\rm{K}]$")
```
