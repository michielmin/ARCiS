```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

```python
# define the directory where the run is stored
dir='/Users/michielm/HAT-P-12b/NIRSPEC_HST/final_TEATRO/'
```

```python
# If there was a retrieval run postprocessed with pew=.true. you can use the trans_limits file
with open(dir+'/trans_limits') as file:
    data1 = [[float(digit) for digit in line.split()] for line in file]

nsamples=len(data1)

mx=[]
mym3=[]
mym2=[]
mym1=[]
myp1=[]
myp2=[]
myp3=[]
my=[]
for i in range(0,nsamples):
    mx.append(data1[i][0])
    my.append(data1[i][4])
    mym3.append(data1[i][1])
    myp3.append(data1[i][7])
    mym1.append(data1[i][3])
    myp1.append(data1[i][5])
    mym2.append(data1[i][2])
    myp2.append(data1[i][6])
```

```python
# Plot the posterior distribution of the transit spectrum
fig, (ax) = plt.subplots(1, 1, sharex=True)

ax.fill_between(mx, mym3, myp3, color=(0,0,1),zorder=120,alpha=0.33)
ax.fill_between(mx, mym2, myp2, color=(0,0,1),zorder=130,alpha=0.33)
ax.fill_between(mx, mym1, myp1, color=(0,0,1),zorder=140,alpha=0.33)
ax.plot(mx, my, color=(0,0,0.5),zorder=150)

ax.set_ylabel(r"$(R_p/R_\star)^2$")
ax.set_xlabel(r"$\lambda [\mu$m]")
```

```python
# to plot a single transit spectrum from a forward model
with open(dir+'/trans_limits') as file:
    data1 = [[digit for digit in line.split()] for line in file]

nsamples=len(data1)

mx=[]
my=[]
for i in range(1,nsamples):
    mx.append(float(data1[i][0]))
    my.append(float(data1[i][1]))
```

```python
# Plot the transit spectrum
fig, (ax) = plt.subplots(1, 1, sharex=True)

ax.plot(mx, my, color=(0,0,0.5),zorder=150)

ax.set_ylabel(r"$(R_p/R_\star)^2$")
ax.set_xlabel(r"$\lambda [\mu$m]")
```
