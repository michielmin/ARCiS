```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
filename='output/cloudstructure0010.dat'

def adjust_lightness(color, amount=0.5):
    import matplotlib.colors as mc
    import colorsys
    try:
        c = mc.cnames[color]
    except:
        c = color
    c = colorsys.rgb_to_hls(*mc.to_rgb(c))
    return colorsys.hls_to_rgb(c[0], max(0, min(1, amount * c[1])), c[2])


with open(filename) as file:
    data1 = [[digit for digit in line.split()] for line in file]

nc=0
while('[s]' in data1[0][nc+4]):
    nc=nc+1

if(nc==0):
    nc=12

print(nc)
names=data1[0][4:4+nc]
print(names)
v_names=data1[0][4+nc:len(data1[0])-5]
print(v_names)

col=[]
for spec in names:
    if(spec=='Al2O3[s]'):
        col.append([1,0,1])
    elif(spec=='TiO2[s]'):
        col.append(adjust_lightness([1,0,1],0.9))
    elif(spec=='Ti4O7[s]'):
        col.append(adjust_lightness([1,0,1],0.8))
    elif(spec=='MgTi2O5[s]'):
        col.append(adjust_lightness([1,0,1],0.7))
    elif(spec=='CaTiO3[s]'):
        col.append(adjust_lightness([1,0,1],0.6))
    elif(spec=='MgO[s]' or spec=='MgO'):
        col.append(adjust_lightness([1,0,1],0.5))
    elif(spec=='MgAl2O4[s]'):
        col.append(adjust_lightness([1,0,1],0.4))

    elif(spec=='FeO[s]'):
        col.append([1,0.67,0.1])
    elif(spec=='Fe2O3[s]'):
        col.append(adjust_lightness([1,0.67,0.1],0.8))
    elif(spec=='Fe3O4[s]'):
        col.append(adjust_lightness([1,0.67,0.1],0.6))

    elif(spec=='Fe[s]' or spec=='Fe'):
        col.append([1,0,0])
    elif(spec=='Zn[s]'):
        col.append(adjust_lightness([1,0,0],0.9))
    elif(spec=='Mn[s]'):
        col.append(adjust_lightness([1,0,0],0.8))
    elif(spec=='Cr[s]'):
        col.append(adjust_lightness([1,0,0],0.7))
    elif(spec=='Ni[s]'):
        col.append(adjust_lightness([1,0,0],0.6))
    elif(spec=='W[s]'):
        col.append(adjust_lightness([1,0,0],0.5))
    elif(spec=='S[s]'):
        col.append(adjust_lightness([1,0,0],0.4))
        
    elif(spec=='SiO[s]' or spec=='SiO'):
        col.append([0,1,0])
    elif(spec=='SiO2[s]' or spec=='Silica'):
        col.append(adjust_lightness([0,1,0],0.9))
    elif(spec=='MgSiO3[s]' or spec=='Pyroxene'):
        col.append(adjust_lightness([0,1,0],0.8))
    elif(spec=='Mg2SiO4[s]' or spec=='Olivine'):
        col.append(adjust_lightness([0,1,0],0.7))
    elif(spec=='CaSiO3[s]'):
        col.append(adjust_lightness([0,1,0],0.6))
    elif(spec=='FeSiO3[s]'):
        col.append(adjust_lightness([0,1,0],0.5))
    elif(spec=='Fe2SiO4[s]'):
        col.append(adjust_lightness([0,1,0],0.4))
    elif(spec=='NaAlSi3O8[s]'):
        col.append(adjust_lightness([0,1,0],0.3))

    elif(spec=='ENSTATITE[s]'):
        col.append(adjust_lightness([0.75,1,0.2],0.8))
    elif(spec=='FORSTERITE[s]'):
        col.append(adjust_lightness([0.75,1,0.2],0.7))
    elif(spec=='FERROSILITE[s]'):
        col.append(adjust_lightness([0.75,1,0.2],0.5))
    elif(spec=='FAYALITE[s]'):
        col.append(adjust_lightness([0.75,1,0.2],0.4))

    elif(spec=='ZnS[s]'):
        col.append([1,0.9,0])
    elif(spec=='FeS[s]'):
        col.append(adjust_lightness([1,0.9,0],0.9))
    elif(spec=='Na2S[s]'):
        col.append(adjust_lightness([1,0.9,0],0.8))
    elif(spec=='MnS[s]'):
        col.append(adjust_lightness([1,0.9,0],0.7))

    elif(spec=='NaCl[s]'):
        col.append([0,0,1])
    elif(spec=='KCl[s]'):
        col.append(adjust_lightness([0,0,1],0.8))
    elif(spec=='NH4Cl[s]'):
        col.append(adjust_lightness([0,0,1],0.6))

    elif(spec=='H2O[s]' or spec=='H2O'):
        col.append([0,1,1])
    elif(spec=='NH3[s]'):
        col.append(adjust_lightness([0,1,1],0.8))
    elif(spec=='NH4SH[s]'):
        col.append(adjust_lightness([0,1,1],0.6))
    elif(spec=='H2SO4[s]'):
        col.append(adjust_lightness([0,1,1],0.4))
    elif(spec=='CH4[s]'):
        col.append(adjust_lightness([0,1,1],0.2))
    else:
        col.append([0.5,0.5,0.5])
```

```python
nsamples=len(data1)

xc=[]
P=[]
xt=[]
r=[]
xv=[]
xn=[]
Js=[]
T=[]
for j in range(1,nsamples):
    P.append(float(data1[j][0]))
    r.append(float(data1[j][-4]))
    T.append(float(data1[j][-3]))
    xn.append(float(data1[j][2]))
    Js.append(float(data1[j][-2]))
    tot=0.0
    for i in range(0,len(names)):
        tot=tot+float(data1[j][i+3])+1e-50
    xt.append(tot)
for i in range(0,len(names)):
    xc.append([])
    for j in range(1,nsamples):
        xc[-1].append(float(data1[j][i+3])+1e-50)
for i in range(0,len(v_names)):
    xv.append([])
    for j in range(1,nsamples):
        xv[-1].append(float(data1[j][i+3+nc]))
print(len(P))
```

```python
minx=1e-10
xf=[]
xf.append([])
for j in range(0,len(P)):
    xf[-1].append(xt[j])
for i in range(0,len(names)):
    xf.append([])
    for j in range(0,len(P)):
        xf[-1].append(10.**(np.log10(xf[-2][j])-max(0.,np.log10(xt[j])-np.log10(minx))*xc[i][j]/xt[j]))
xf.append([])
for j in range(0,len(P)):
    xf[-1].append(minx)

for i in range(1,len(names)+1):
#    plt.plot(xc[i],P,label=names[i])
    plt.fill_betweenx(P,xf[i-1],xf[i],label=names[i-1],color=col[i-1],linewidth=0)
#plt.plot(xt,P,linewidth=4)
plt.plot(np.array(r)/1e2,P,label='radius',color='black',linewidth=2)

#plt.plot(np.array(Js)*1e6,P,label='Jstar',color='green',linewidth=5,zorder=500)

for i in range(0,len(v_names)):
    ls='-'
    if(i >= 10):
        ls='--'
    plt.plot(xv[i],P,label=v_names[i],linestyle=ls,linewidth=2)

#i=2
#plt.plot(xv[i],P,label=v_names[i])

#plt.plot(xn,P,label='xn',linewidth=2)

plt.xscale('log')
plt.yscale('log')
plt.xlabel('mass fraction / particle size [m]')
plt.ylabel('pressure [bar]')
plt.ylim(1e1,1e-8)
plt.xlim(minx,1e0)
plt.legend(loc='upper center', bbox_to_anchor=(1.35, 1.15),
          ncol=3, fancybox=True, shadow=True)
```

```python
xfrac=[]
xfrac.append([])
for j in range(0,len(P)):
    if(xt[j] > 1e-40):
        xfrac[-1].append(0.0)
    else:
        xfrac[-1].append(1.0)
for i in range(0,len(names)):
    xfrac.append([])
    for j in range(0,len(P)):
        if(xt[j] > 1e-40):
            xfrac[-1].append(xfrac[-2][j]+xc[i][j]/xt[j])
        else:
            xfrac[-1].append(1.0)
xfrac.append([])
for j in range(0,len(P)):
    xfrac[-1].append(1.0)

for i in range(1,len(names)+1):
    plt.fill_betweenx(P,xfrac[i-1],xfrac[i],label=names[i-1],color=col[i-1],linewidth=0)

plt.yscale('log')
plt.xlabel('mass fraction')
plt.ylabel('pressure [bar]')
plt.ylim(1e3,1e-12)
plt.xlim(0,1)
plt.legend(loc='upper center', bbox_to_anchor=(1.35, 1.15),
          ncol=3, fancybox=True, shadow=True)
```
